```python
import numpy as np
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
X = np.random.uniform(size=(10000, 10))
```

```python
pca = PCA(n_components=2)
```

```python
Xt = pca.fit_transform(X)
```

```python
plt.scatter(Xt[:, 0], Xt[:, 1])
```
